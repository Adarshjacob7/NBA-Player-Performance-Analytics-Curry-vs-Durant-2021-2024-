```python
# 2023-2024 (same code is used for 2021, 2022 & 2024 seasons)
```

```python
from nba_api.stats.endpoints import playergamelog
from nba_api.stats.static import players
import pandas as pd

# -------- Helper: Get player ID by name -------- #
def get_player_id(player_name):
    all_players = players.get_players()
    for p in all_players:
        if p['full_name'].lower() == player_name.lower():
            return p['id']
    return None

# -------- Step 1: Get player IDs -------- #
curry_id = get_player_id("Stephen Curry")
durant_id = get_player_id("Kevin Durant")

print("Curry ID:", curry_id)
print("Durant ID:", durant_id)

# -------- Step 2: Fetch 2024 game logs -------- #
curry_logs = playergamelog.PlayerGameLog(player_id=curry_id, season="2023-24").get_data_frames()[0]
durant_logs = playergamelog.PlayerGameLog(player_id=durant_id, season="2023-24").get_data_frames()[0]

# -------- Step 3: Save to CSV -------- #
curry_logs.to_csv("curry_2023_gamelog.csv", index=False)
durant_logs.to_csv("durant_2023_gamelog.csv", index=False)

print("Completed! CSV files saved.")
```
